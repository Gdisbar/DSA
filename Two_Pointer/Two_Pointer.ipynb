## 838. Push Dominoes

There are n dominoes in a line, and we place each domino vertically upright. In the beginning, we simultaneously push some of the dominoes either to the left or to the right.

After each second, each domino that is falling to the left pushes the adjacent domino on the left. Similarly, the dominoes falling to the right push their adjacent dominoes standing on the right.

When a vertical domino has dominoes falling on it from both sides, it stays still due to the balance of the forces.

For the purposes of this question, we will consider that a falling domino expends no additional force to a falling or already fallen domino.

You are given a string dominoes representing the initial state where:

dominoes[i] = 'L', if the ith domino has been pushed to the left,
dominoes[i] = 'R', if the ith domino has been pushed to the right, and
dominoes[i] = '.', if the ith domino has not been pushed.
Return a string representing the final state.

 

        Example 1:

        Input: dominoes = "RR.L"
        Output: "RR.L"
        Explanation: The first domino expends no additional force on the second domino.
        Example 2:


        Input: dominoes = ".L.R...LR..L.."
        Output: "LL.RR.LLRRLL.."
 

```python

Approach-1 (Two pointer)
-------------------------------------------------
In this approach, you just need to find sections like this
X .   .   .   . X
i                j
Where X can be 'R' or 'L' and in between there can be as many dots
Now,
- you know the length of mid part
- If char[i] == char[j] == 'R', means all go towards right (R)
-  char[i]  == char[j] == 'L', means all go towards Left (L)
-  If char[i] = 'L' and char[j] = 'R', means middle part is not affected so the remain '.'
-  If char[i] = 'R' and char[j] = 'L', then it will affect the middle part.
    The middle_part/2 close to i will be affected by 'R' and middle_part/2 close to j will be   
    effected by 'L'  and the last mid point (middle_part%2) will be unaffected due to equal  
    force from left and right so it remains '.'


def pushDominoes(self, dominoes: str) -> str:
    s = "L" + dominoes + "R" # avoids having to handle boundary cases separately.
    n = len(s)
    result = ""
    i = 0
    for j in range(1,n):
        if s[j]==".": # no force yet
            continue
        # number of dominoes between boundary s[i] and s[j]
        midPart = j-i-1 
        # append the left boundary character (skip the dummy 'L' at the start).
        if i>0: 
            result+=s[i]
        # Same direction (L...L or R...R) → all middle dominoes fall the same way.
        if s[i]==s[j]:
            result+=s[i]*midPart
        # Opposite but outward forces (L...R) → middle dominoes remain upright ('.').
        elif s[i]=='L' and s[j]=='R': 
            result+='.'*midPart
        else:
            result+='R'*(midPart//2)+'.'*(midPart%2)+'L'*(midPart//2)
        i=j
    return result


Approach-2 (Using 2 different array)
-------------------------------------------------
def pushDominoes(self, dominoes: str) -> str:
    n = len(dominoes)
    rightClosestL = [n] * n # n as the sentinel (past the end)
    leftClosestR = [-1] * n # -1 as the sentinel (before the start)
    # rightClosestL -> push from right to left,closest 'L' to the right
    for i in range(n-1,-1,-1):
        if dominoes[i]=='L':
            rightClosestL[i]=i
        elif dominoes[i]=='.':
            if i < n-1: # propagate the index from the right
                rightClosestL[i]= rightClosestL[i+1]
        
    # leftClosestR -> push from left to right,closest 'R' to the left
    for i in range(0,n):
        if dominoes[i]=='R':
            leftClosestR[i]=i
        elif dominoes[i]=='.':
            if i>0: # Propagate the index from the left
                leftClosestR[i]=leftClosestR[i-1]
        
    
    result = [''] * n
    for i in range(n):
        # If the domino is already 'L' or 'R', keep it
        if dominoes[i] != '.':
            result[i] = dominoes[i]
            continue

        # Distance to closest L on the right (or n if not found)
        distL = rightClosestL[i] - i 
        # Distance to closest R on the left (or i - (-1) = i+1 if not found)
        distR = i - leftClosestR[i]
        
        # Use the sentinel values (n and -1) for comparison
        is_pushed_L = rightClosestL[i] != n
        is_pushed_R = leftClosestR[i] != -1
        
        if is_pushed_L and is_pushed_R:
            # Pushed from both sides
            if distL < distR:
                result[i] = 'L'
            elif distL > distR:
                result[i] = 'R'
            else: # distL == distR
                result[i] = '.'
        elif is_pushed_L:
            # Pushed only from the right by 'L'
            result[i] = 'L'
        elif is_pushed_R:
            # Pushed only from the left by 'R'
            result[i] = 'R'
        else:
            # Not pushed by either 'L' or 'R'
            result[i] = '.'
                
    return ''.join(result)
        


```
