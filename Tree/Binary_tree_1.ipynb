## 105. Construct Binary Tree from Preorder and Inorder Traversal

Input: preorder = [3,9,20,15,7], inorder = [9,3,15,20,7]
Output: [3,9,20,null,null,15,7]
Example 2:

Input: preorder = [-1], inorder = [-1]
Output: [-1]

```python

def solve(self,preorder,inorder,start,end,idx):
        if start > end:
            return None
        root_val = preorder[idx[0]]
        root = TreeNode( root_val)
        idx[0]+=1
        mid = inorder.index(root_val)
        root.left = self.solve(preorder,inorder,start,mid-1,idx)
        root.right = self.solve(preorder,inorder,mid+1,end,idx)
        
        return root
def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
    n = len(preorder)
    return self.solve(preorder,inorder,0,n-1,[0])

```

## 236. Lowest Common Ancestor of a Binary Tree

    Input: root = [3,5,1,6,2,0,8,null,null,7,4], p = 5, q = 1
    Output: 3
    Explanation: The LCA of nodes 5 and 1 is 3.


    Input: root = [3,5,1,6,2,0,8,null,null,7,4], p = 5, q = 4
    Output: 5
    Explanation: The LCA of nodes 5 and 4 is 5, since a node can be a descendant of itself according to the LCA definition.

```python

"""
O(N^2) Approach
"""

    def solve(self,root,p_q):
        if not root: return
        if root==p_q:
            return True
        # Check if we can find a 'p'/'q' in deeper tree
        if root.left:
            if self.solveP(root.left,p_q):
                return True
        if root.right:
            if self.solveP(root.right,p_q):
                return True
        
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if not root: return None
        findP = self.solve(root,p)
        findQ = self.solve(root,q)
        if findP and findQ:
            """
            Check if there is a "deeper" LCA in the children first
            similar to solveP,solveQ but since we need value(node) 
            we store them in variable initialized inside condition
            that we already have a common ancestor but we need a 
            lower one"""
            res_left = None
            res_right = None
            if root.left:
                res_left = self.lowestCommonAncestor(root.left, p, q)
            if res_left: return res_left
            
            if root.right:
                res_right = self.lowestCommonAncestor(root.right, p, q)
            if res_right: return res_right
            
            # If neither child is an LCA, then the current root MUST be the LCA
            return root


"""
Efficient - O(N)
"""

def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':

        # Base case: if we hit a null or find one of the targets
        if not root or root == p or root == q:
            return root
        
        # Look for p and q in the left and right subtrees
        left = self.lowestCommonAncestor(root.left, p, q)
        right = self.lowestCommonAncestor(root.right, p, q)
        
        # If both sides returned a node, this root is the LCA
        if left and right:
            return root
        
        # Otherwise, return the one that wasn't None (if any)
        return left if left else right
```
