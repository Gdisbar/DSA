    s contains English letters (upper-case and lower-case), digits, and spaces ' '.

```python

"  elpmaxe   doog a" # after reversing

def reverse_words(self,a,n):
  i,j = 0,0
  while i < n :
      while i < j or i < n and a[i]==' ': # skip space
              i+=1
      while j < i or j < n and a[j]!=' ': # skip letters
              j+=1
      self.reverse_str(a,i,j-1)

i = 2  # after skipping spaces -> a[i==2]='e'
j = 9 # after skipping letters -> a[j==9]=' '
reverse_str(a, 2, 8)

i = 2 (not < j → 2 < 9 → yes, so increment i until i >= j) → i catches up to j=9

Then a[i==9] == ' ' so we skip more.
✅ Now i = 12, j = 9

similarly j catch-up to i=12 then skip letters & j = 17
reverse_str(a, 12, 16)

then i = 17,j = 18
reverse_str(a, 17, 17)
```
