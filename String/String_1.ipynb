## 38. Count and Say

The count-and-say sequence is a sequence of digit strings defined by the recursive formula:

countAndSay(1) = "1"
countAndSay(n) is the run-length encoding of countAndSay(n - 1).

    Example 1:

    Input: n = 4

    Output: "1211"

    Explanation:

    countAndSay(1) = "1"
    countAndSay(2) = RLE of "1" = "11"
    countAndSay(3) = RLE of "11" = "21"

```python
def helper(n):
    if n==1: return "1"
    
    say = helper(n-1) # get the string for n-1
    result = ""
    i = 0
    while i < len(say):
        cur_char = say[i]
        count = 0
        while i < len(say) and say[i]==cur_char:
            count+=1
            i+=1
        result+=str(count)+cur_char

    return result
```
