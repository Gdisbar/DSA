````python
def countAndSay(self, n: int) -> str:
        if n==1: return "1"
        say = self.countAndSay(n-1)
        # process
        i,cnt = 0,1
        prev_term = say[i]
        next_term = ""
        print(f"recur={n-1}, say = {say},cnt={cnt},prev_term={prev_term}")
        while i < len(say) and prev_term==say[i]:
            cnt+=1
            i+=1
        print(f"recur={n-1}, i={i},cnt={cnt}")
        if cnt>1:
            next_term+=str(cnt)+prev_term
        return next_term
````

````python
class Solution:
    def countAndSay(self, n: int) -> str:
        
        def helper(n):
            if n==1: return "1"
            
            say = helper(n-1) # get the string for n-1
            result = ""
            i = 0
            while i < len(say):
                j = i
                while j < len(say) and say[j]==say[i]:
                    j+=1
                result+=str(j-i)+say[i]
                i=j
            return result
        
        return helper(n)
            

        
````
