## 823. Binary Trees With Factors

Given an array of unique integers, arr, where each integer arr[i] is strictly greater than 1.
We make a binary tree using these integers, and each number may be used for any number of times. Each non-leaf node's value should be equal to the product of the values of its children.

Return the number of binary trees we can make. The answer may be too large so return the answer modulo 10^9 + 7.

 

    Example 1:

    Input: arr = [2,4]
    Output: 3
    Explanation: We can make these trees: [2], [4], [4, 2, 2]
    Example 2:

    Input: arr = [2,4,5,10]
    Output: 7
    Explanation: We can make these trees: [2], [4], [5], [10], [4, 2, 2], [10, 2, 5], [10, 5, 2].
    

    Constraints:

    1 <= arr.length <= 1000
    2 <= arr[i] <= 109
    All the values of arr are unique.

```python
       0 1 2
arr = [2,4,8]

mp = {
    2 : 1,
    4: 1,
    8:1
}

i = 1, j = 0 to 0
-------------
j = 0, arr[j]=2 -> pair exist [arr[j]=2,arr[i]/arr[j]=2]
mp = {
    4: 1+1
}

i = 2, j = 0 to 1
-------------------
j = 0, arr[j]=2 -> pair exist [arr[j]=2,arr[i]/arr[j]=4]
mp = {
    8: 1+1
}
j = 1, arr[j]=4 -> pair exist [arr[j]=4,arr[i]/arr[j]=2]
mp = {
    8:1+1+1
}

mp = {
    2:1,
    4:1+1,
    8:1+1+1
}
```

```python

def numFactoredBinaryTrees(self, arr: List[int]) -> int:
    mp = {}
    modulo = 10**9+7
    arr = sorted(arr)
    n = len(arr)
    for i in range(n):
        mp[arr[i]] = 1
    
    for i in range(1,n):
        for j in range(0,i):
            if i==j:
                continue
            else:
                if arr[i]%arr[j]==0:
                    x = arr[i]/arr[j]
                    if x in mp:
                        right_child = mp[x]
                        left_child = mp[arr[j]]
                        total_ways = (left_child*right_child)%modulo
                        mp[arr[i]] = mp[arr[i]]+total_ways
    
    result = 0
    print(mp)
    for vals in mp.values():
        result+=vals
        result = result % modulo
    return result

        

```

```python


from collections import defaultdict

class Solution:
    def dfs(self,node,visited,adj_list,result):
        if node not in visited:
            visited.add(node)
            result.append(node)
        for neighbour in adj_list[node]:
            if neighbour not in visited:
                self.dfs(neighbour,visited,adj_list,result)

    def restoreArray(self, adjacentPairs: List[List[int]]) -> List[int]:
        adj_list = defaultdict(list)
        
        for pair in adjacentPairs:
            u,v = pair
            adj_list[u].append(v)
            adj_list[v].append(u)
        
        start_node = float('inf')
        for key,values in adj_list.items():
            if len(values)==1:
                start_node = key
                break
        
        visited = set()
        result = []
        print(adj_list)
        print(start_node)
        self.dfs(start_node,visited,adj_list,result)
        return result

```
