## Factorials of large numbers

```python

def factorial(self, n):
    nums = [0 for _ in range(3000)] # storing factorial upto 1000 safely
    nums[0]=1
    res_size = 1
    def multiply(x:int):
        # nonlocal - variable defined in an outer (parent) function
        # global - variable defined in top level of a file/module
        nonlocal res_size 
        carry = 0
        for i in range(res_size):
            result = x*nums[i]+carry
            nums[i]=result%10
            carry = result//10
        while carry>0:
            nums[res_size]=carry%10
            carry=carry//10
            res_size+=1

    for i in range(2,n+1):
        multiply(i)
    
    result = []
    for i in range(res_size-1,-1,-1):
        result.append(nums[i])
    return result

```
